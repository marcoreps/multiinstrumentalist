```python
import numpy as np
import matplotlib.pyplot as plt

# Array from .csv
arr=np.genfromtxt('https://raw.githubusercontent.com/marcoreps/multiinstrumentalist/master/csv/20220911-072449_REPS5700A_PTB3458A_3458A_3458B_INL.csv',delimiter=',',skip_header=1)

F5700A = [col[0] for col in arr]
HP3458A = [col[1] for col in arr]
HP3458B = [col[2] for col in arr]
HP3458P = [col[3] for col in arr]
PPMdeviationA = [(F5700A[x]-HP3458A[x])/(10/1000000) for x in range(len(HP3458A))]
PPMdeviationB = [(F5700A[x]-HP3458B[x])/(10/1000000) for x in range(len(HP3458A))]
PPMdeviationP = [(F5700A[x]-HP3458P[x])/(10/1000000) for x in range(len(HP3458A))]

figa, axa = plt.subplots()
#axa.plot(F5700A, PPMdeviationa, label='absolute')
axa.set_xlabel('Calibrator Vout')
axa.set_ylabel('PPM of 10V range')
axa.set_title("REPS 5700A vs 3x3458A INL test")
figa.set_size_inches(16, 9)
figa.patch.set_facecolor('white')

# Fit a polynomial function of degree 1
termsA = np.polyfit(F5700A, PPMdeviationA, 1)
termsB = np.polyfit(F5700A, PPMdeviationB, 1)
termsP = np.polyfit(F5700A, PPMdeviationP, 1)

# Create a function with the terms we just generated
polyfunctionA = np.poly1d(termsA)
polyfunctionB = np.poly1d(termsB)
polyfunctionP = np.poly1d(termsP)

bestfitA = [PPMdeviationA[x]-polyfunctionA(F5700A[x]) for x in range(len(HP3458A))]
bestfitB = [PPMdeviationB[x]-polyfunctionB(F5700A[x]) for x in range(len(HP3458B))]
bestfitP = [PPMdeviationP[x]-polyfunctionP(F5700A[x]) for x in range(len(HP3458P))]

# Plot a little test
axa.plot(F5700A, bestfitA, label='Reps 3458A')
axa.plot(F5700A, bestfitB, label='Reps 3458B')
axa.plot(F5700A, bestfitP, label='PTB 3458A')
axa.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Array from .csv
arr=np.genfromtxt('https://raw.githubusercontent.com/marcoreps/multiinstrumentalist/master/csv/20220911-065528_PTB5700A_PTB3458A_3458A_3458B_INL.csv',delimiter=',',skip_header=1)

F5700A = [col[0] for col in arr]
HP3458A = [col[1] for col in arr]
HP3458B = [col[2] for col in arr]
HP3458P = [col[3] for col in arr]
PPMdeviationA = [(F5700A[x]-HP3458A[x])/(10/1000000) for x in range(len(HP3458A))]
PPMdeviationB = [(F5700A[x]-HP3458B[x])/(10/1000000) for x in range(len(HP3458A))]
PPMdeviationP = [(F5700A[x]-HP3458P[x])/(10/1000000) for x in range(len(HP3458A))]

F5700Aneg = F5700A[:len(F5700A)//2]
F5700Apos = F5700A[len(F5700A)//2:]

PPMdeviationAneg = PPMdeviationA[:len(PPMdeviationA)//2]
PPMdeviationApos = PPMdeviationA[len(PPMdeviationA)//2:]

PPMdeviationBneg = PPMdeviationB[:len(PPMdeviationB)//2]
PPMdeviationBpos = PPMdeviationB[len(PPMdeviationB)//2:]

PPMdeviationPneg = PPMdeviationP[:len(PPMdeviationP)//2]
PPMdeviationPpos = PPMdeviationP[len(PPMdeviationP)//2:]

figa, axa = plt.subplots()
#axa.plot(F5700A, PPMdeviationa, label='absolute')
axa.set_xlabel('Calibrator Vout')
axa.set_ylabel('PPM of 10V range')
axa.set_title("PTB 5700A vs 3x3458A INL test")
figa.set_size_inches(16, 9)
figa.patch.set_facecolor('white')

# Fit a polynomial function of degree 1
termsAneg = np.polyfit(F5700Aneg, PPMdeviationAneg, 1)
termsApos = np.polyfit(F5700Apos, PPMdeviationApos, 1)

termsBneg = np.polyfit(F5700Aneg, PPMdeviationBneg, 1)
termsBpos = np.polyfit(F5700Apos, PPMdeviationBpos, 1)

termsPneg = np.polyfit(F5700Aneg, PPMdeviationPneg, 1)
termsPpos = np.polyfit(F5700Apos, PPMdeviationPpos, 1)

# Create a function with the terms we just generated
polyfunctionAneg = np.poly1d(termsAneg)
polyfunctionApos = np.poly1d(termsApos)
polyfunctionBneg = np.poly1d(termsBneg)
polyfunctionBpos = np.poly1d(termsBpos)
polyfunctionPneg = np.poly1d(termsPneg)
polyfunctionPpos = np.poly1d(termsPpos)

bestfitAneg = [PPMdeviationAneg[x]-polyfunctionAneg(F5700Aneg[x]) for x in range(len(PPMdeviationAneg))]
bestfitApos = [PPMdeviationApos[x]-polyfunctionApos(F5700Apos[x]) for x in range(len(PPMdeviationApos))]

bestfitBneg = [PPMdeviationBneg[x]-polyfunctionBneg(F5700Aneg[x]) for x in range(len(PPMdeviationBneg))]
bestfitBpos = [PPMdeviationBpos[x]-polyfunctionBpos(F5700Apos[x]) for x in range(len(PPMdeviationBpos))]

bestfitPneg = [PPMdeviationPneg[x]-polyfunctionPneg(F5700Aneg[x]) for x in range(len(PPMdeviationPneg))]
bestfitPpos = [PPMdeviationPpos[x]-polyfunctionPpos(F5700Apos[x]) for x in range(len(PPMdeviationPpos))]

# Plot a little test
axa.plot(F5700A, bestfitAneg+bestfitApos, label='Reps 3458A')
axa.plot(F5700A, bestfitBneg+bestfitBpos, label='Reps 3458B')
axa.plot(F5700A, bestfitPneg+bestfitPpos, label='PTB 3458A')
axa.legend()
```
